```
class Node:
  def __init__(self,predicted_class,genie,Entropy):
    self.prediction=predicted_class
    self.Entropy=1
    self.genie=1
    self.col_idx=0
    self.threshold=0
    self.left=None
    self.right=None

def fit(X,y,max_depth,split_metric):
  tree=make_tree(X,y,0,max_depth,split_metric)
  return tree
def Entropy_parent(left):
  if left[0]==0 or left[1]==0:
    return 0
  total_left=left[0]+left[1]
  if total_left==0:
    return 0
  Eleft=0-((float(left[0])/total_left)*math.log2(float(left[0])/total_left)+(float(left[1])/total_left)*math.log2(float(left[1])/total_left))
  return Eleft

def Entropy(left,right):
  total_left=left[0]+left[1]
  total_right=right[0]+right[1]
  if left[0]==0 or left[1]==0:
    if right[0]==0 or right[1]==0:
      return 0
    else:
      Eright=0-((float(right[0])/total_right)*math.log2(float(right[0])/total_right)+(float(right[1])/total_right)*math.log2(float(right[1])/total_right))
      return Eright
  if right[0]==0 or right[1]==0:
    if left[0]==0 or left[1]==0:
      return 0
    else:
      Eleft=0-((float(left[0])/total_left)*math.log2(float(left[0])/total_left)+(float(left[1])/total_left)*math.log2(float(left[1])/total_left))
      return Eleft
  Eleft=0-((float(left[0])/total_left)*math.log2(float(left[0])/total_left)+(float(left[1])/total_left)*math.log2(float(left[1])/total_left))
  if total_right==0:
    return gleft
  Eright=0-((float(right[0])/total_right)*math.log2(float(right[0])/total_right)+(float(right[1])/total_right)*math.log2(float(right[1])/total_right))
  total=total_left+total_right
  E=(float(total_left)/total)*Eleft+(float(total_right)/total)*Eright
  return E
def make_tree(X,y,depth,max_depth,split_metric):
  nfeatures=X.shape[1]
  classes=[2,4]
  number_of_elements=[]
  for ele in classes:
    count=0
    for i in range(len(y)):
      if y[i]==ele:
        count+=1
    number_of_elements.append(count)
    
  max=0
  index=0
  for i in range(len(number_of_elements)):
    if(number_of_elements[i]>max):
      max=number_of_elements[i]
      index=i
  g=0
  node=Node(classes[index],1,1)
  if split_metric=="genie":
    g=genie_parent(number_of_elements)
    node.genie=g
  elif split_metric=="Entropy":
    g=Entropy_parent(number_of_elements)
    node.Entropy=g
  
  if g==0.0:
      return node
  if depth<max_depth:
    thres,indx = best_split(X,y,nfeatures,split_metric)
    print("col_index:",indx,"thres:",thres)
    if indx!=None:
      left_indices=X[:,indx]<=thres
      right_indices=X[:,indx]>thres
      X_left,y_left=X[left_indices],y[left_indices]
      X_right,y_right=X[right_indices],y[right_indices]
      node.col_idx=indx
      node.threshold=thres
      #print("col_index:",indx,"thres:",thres)
      node.left=make_tree(X_left,y_left,depth+1,max_depth,split_metric)
      node.right=make_tree(X_right,y_right,depth+1,max_depth,split_metric)
  return node
def predictionHelper(ele,root):
  while root!=None:
    index=root.col_idx
    if ele[index]<=root.threshold:
      if root.left!=None:
        root=root.left
      else:
        return root.prediction
    else:
      if root.right!=None:
        root=root.right
      else:
        return root.prediction
def genie_parent(left):
  #print(left)
  total_left=left[0]+left[1]
  if total_left==0:
    return 0
  gleft=1-((float(left[0])/total_left)**2+(float(left[1])/total_left)**2)
  return gleft
def predict(X,root):
  result=[]
  for ele in X:
    result.append(predictionHelper(ele,root))
  return result
def genie(left,right):
  total_left=left[0]+left[1]
  total_right=right[0]+right[1]
  gleft=1-((float(left[0])/total_left)**2+(float(left[1])/total_left)**2)
  if total_right==0:
    return gleft
  gright=1-((float(right[0])/total_right)**2+(float(right[1])/total_right)**2)
  total=total_left+total_right
  g=(float(total_left)/total)*gleft+(float(total_right)/total)*gright
  return g
def best_split(X,y,numfeatures,split_metric):
  best_thres=0
  best_gen=9999
  best_feature=99
  for col in range(numfeatures):
    temp=[]
    class_ben=0 #2
    class_malig=0 #4
    for i in range(len(X)):
      x=[]      
      x.append(X[i][col])
      x.append(y[i])
      if y[i]==2:
        class_ben+=1
      temp.append(x)
    class_malig=len(temp)-class_ben
    left_tree=[0,0]
    right_tree=[class_ben,class_malig]
    temp.sort(key=lambda x:x[0])
    i=0
    while i <(len(temp)):
      thres=temp[i][0]
      while i<len(temp) and temp[i][0]==thres:
        if temp[i][1]==2:
          right_tree[0]-=1
          left_tree[0]+=1
        elif temp[i][1]==4:
          right_tree[1]-=1
          left_tree[1]+=1
        i+=1
      sm=0
      if split_metric=="genie": 
        sm=genie(left_tree,right_tree)
      elif split_metric=="Entropy":
        sm=Entropy(left_tree,right_tree)
      if sm<best_gen:
        best_gen=sm
        best_thres=thres
        best_feature=col
  return best_thres,best_feature

```
